```python
import copy
import time
from contextlib import contextmanager
from btree import BinaryTree
```

```python
@contextmanager
def timed_operation():
    start = time.time()
    yield
    end = time.time()
    print(f"The operation took {end - start:.7f} seconds.")
```

```python
binary_tree = BinaryTree(1)

binary_tree.add_left_child(2)
binary_tree.add_right_child(3)

binary_tree.left_child.add_left_child(4)
binary_tree.left_child.add_right_child(5)

binary_tree.right_child.add_left_child(6)
binary_tree.right_child.add_right_child(7)

binary_tree.left_child.left_child.add_left_child(8)
binary_tree.left_child.left_child.add_right_child(9)

binary_tree.left_child.right_child.add_left_child(10)
binary_tree.left_child.right_child.add_right_child(11)

binary_tree.right_child.left_child.add_left_child(12)
binary_tree.right_child.left_child.add_right_child(13)

binary_tree.right_child.right_child.add_left_child(14)
binary_tree.right_child.right_child.add_right_child(15)
```

```python
print("Original Tree:")
binary_tree.visualize()
```

```python
binary_tree_bfs = copy.deepcopy(binary_tree)

with timed_operation():
    # Invert the binary tree
    binary_tree_bfs.invert_dfs()

print("\nInverted Tree:")
binary_tree_bfs.visualize(color='b')
```

```python
binary_tree_dfs = copy.deepcopy(binary_tree)

with timed_operation():
    # Invert the binary tree
    binary_tree_dfs.invert_dfs()

print("\nInverted Tree:")
binary_tree_dfs.visualize(color='r')
```
